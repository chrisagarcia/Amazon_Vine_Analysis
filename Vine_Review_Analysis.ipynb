```python
import pandas as pd
```

#### the next few cells require the tsv.gz file that is way too big to upload to github but the link is there if you want to download

```python
file = pd.read_csv("amazon_reviews_us_Lawn_and_Garden_v1_00.tsv.gz", sep="\t", nrows=7905)
```

```python
df = pd.DataFrame(file)
```

```python
file_2 = pd.read_csv("amazon_reviews_us_Lawn_and_Garden_v1_00.tsv.gz", sep="\t", names=df.columns, skiprows=7905, )
```

```python
df_2 = pd.DataFrame(file_2)
```

```python
merge_df = df.append(df_2)
```

```python
merge_df
```

```python
# step 1

vote_filter = merge_df[merge_df["total_votes"] >= 20]
vote_filter.head()
```

```python
# step 2

good_reviews = vote_filter[(vote_filter["helpful_votes"] / vote_filter["total_votes"]) >= .5]
good_reviews.head(5)
```

```python
# step 3

vine_reviews = good_reviews[good_reviews["vine"] == "Y"]

# step 4

non_vine = good_reviews[good_reviews["vine"] == "N"]
```

```python
vine_stats = {
    "5-star reviews": vine_reviews[vine_reviews["star_rating"] == 5]["star_rating"].count(),
    "percentage of 5-star": vine_reviews[vine_reviews["star_rating"] == 5]["star_rating"].count() / vine_reviews["star_rating"].count()
}

natural_stats = {
    "5-star reviews": non_vine[non_vine["star_rating"] == 5]["star_rating"].count(),
    "percentage of 5-star": non_vine[non_vine["star_rating"] == 5]["star_rating"].count() / non_vine["star_rating"].count()
}

print(vine_stats)
print(natural_stats)
```

```python
result = pd.DataFrame([{'5-star reviews': 176, 'percentage of 5-star': 0.45595854922279794}, {'5-star reviews': 24016, 'percentage of 5-star': 0.49315181009877}], index=["vine", "non-vine"])
```
